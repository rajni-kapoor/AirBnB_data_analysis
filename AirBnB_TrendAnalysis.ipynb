```python
# Drop useless column 
df_2020 = df_2020.drop(columns=["neighbourhood_group"], errors="ignore") 

# Fill reviews_per_month 
df_2020["reviews_per_month"] = df_2020["reviews_per_month"].fillna(0) 

# Fill host_name if you want 
df_2020["host_name"] = df_2020["host_name"].fillna("Unknown")

# Converting the datatype of last_review to datetime
df_2020["last_review"] = pd.to_datetime(df_2020["last_review"], errors="coerce")
```

```python
# Drop useless column 
df_2023 = df_2023.drop(columns=["neighbourhood_group"], errors="ignore") 

# Fill reviews_per_month 
df_2023["reviews_per_month"] = df_2023["reviews_per_month"].fillna(0) 

# Fill host_name if you want 
df_2023["host_name"] = df_2023["host_name"].fillna("Unknown")

# Converting the datatype of last_review to datetime
df_2023["last_review"] = pd.to_datetime(df_2023["last_review"], errors="coerce")
```

```python
# adding a year column in the both dataframes 
df_2020["year"] = 2020
df_2023["year"] = 2023
```

```python
# Removing the records where the price is 0 which means the property is unavailable for booking
import numpy as np

df_2020["price"] = df_2020["price"].replace(0, np.nan)
df_2023["price"] = df_2023["price"].replace(0, np.nan)
```

```python
# Removing or capping the values in minimum_nights column so it doesnt skew our analysis of minimumnight trends(some hosts enter huge numbers to block bookings without officially pausing their listing)
df_2020 = df_2020[df_2020["minimum_nights"] < 365]
df_2023 = df_2023[df_2023["minimum_nights"] < 365]
```

```python
# Creating a column demand_category from reviews_per_month column

df_2020["demand_category"] = pd.cut( df_2020["reviews_per_month"], 
                                    bins=[-0.01, 0, 0.3, 1, 2, df_2020["reviews_per_month"].max()], 
                                    labels=["No demand", "Low", "Medium", "High", "Very high"] ) 
df_2023["demand_category"] = pd.cut( df_2023["reviews_per_month"], 
                                    bins=[-0.01, 0, 0.3, 1, 2, df_2023["reviews_per_month"].max()], 
                                    labels=["No demand", "Low", "Medium", "High", "Very high"] )
```
